```python
def num_plotter(data, target):
    for col in data.select_dtypes("number"):
        sns.boxplot(data=data, x=target, y=col)
        plt.show()

num_plotter(bank_df, "Exited")
```

```python
def cat_plotter(data, target):
    for col in data.select_dtypes("object").columns[1:]:
        sns.barplot(data=data, x=target, y=col)
        plt.show()

cat_plotter(bank_df, "Exited")
```
